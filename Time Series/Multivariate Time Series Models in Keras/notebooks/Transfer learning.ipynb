```python
def mape(y_true, y_pred):
    import keras.backend as K
    """
    Returns the mean absolute percentage error.
    For examples on losses see:
    https://github.com/keras-team/keras/blob/master/keras/losses.py
    """
    return (K.abs(y_true - y_pred) / K.abs(y_pred)) * 100

def smape(y_true, y_pred):
    import keras.backend as K
    """
    Returns the Symmetric mean absolute percentage error.
    For examples on losses see:
    https://github.com/keras-team/keras/blob/master/keras/losses.py
    """
    return (K.abs(y_pred - y_true) / ((K.abs(y_true) + K.abs(y_pred))))*100
```

```python
from keras.models import Model
from keras.applications.densenet import preprocess_input, DenseNet121
from keras.layers import Reshape, CuDNNGRU

image_size = (120, 60)

batch_size = 8
num_epochs = 20
weights_final = 'model-DenseNet121-final.h5'

net = DenseNet121(include_top=False, weights='imagenet', input_tensor=None,
               input_shape=(image_size[0], image_size[1], 3))
```

```python
x = net.output

# Add the following in chronologial order after net.output
output_shape = net.output_shape
print(output_shape)

x = Reshape((output_shape[1], output_shape[2]*output_shape[3]))(x)

# Input should be: (batch_size, timesteps, input_dim)
x = CuDNNGRU(16, return_sequences=True, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

x = CuDNNGRU(16, return_sequences=True, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

x = CuDNNGRU(16, return_sequences=False, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

x = Dense(128, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

x = Dense(64, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

x = Dense(16, kernel_initializer='TruncatedNormal')(x)
x = LeakyReLU()(x)
x = Dropout(0.2)(x)

output_layer = Dense(1)(x)

net_final = Model(inputs=net.input, outputs=output_layer)
```

```python
freeze_layers = 15

for layer in net_final.layers[:freeze_layers]:
    layer.trainable = False
for layer in net_final.layers[freeze_layers:]:
    layer.trainable = True

    
#multi_model = multi_gpu_model(net_final, gpus=num_gpu)

net_final.compile(optimizer=Adam(lr=1e-5), # 1e-5
                  loss='mse', metrics=[mape, smape])
```

```python
# Calculate evaluation metrics over the result

ytrue = result['y_true']
ypred = result['y_pred']
n = len(result)

mse_result = (1/n)*np.sum((ypred - ytrue)**2)
mape_result = (100/n) * np.sum(np.abs((ytrue - ypred) / ypred))
smape_result = (100/n) * np.sum( np.abs((ytrue - ypred)) / (np.abs(ytrue) + np.abs(ypred)) )
```
